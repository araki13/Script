```python
#Import the required libraries.
import pandas as pd
import os
import glob
import datetime
from moviepy.editor import *
```

```python
# Change the directory where the CSV file is located.
os.chdir('C:/Users/51623/Desktop/temp/データ分析/データ抽出/元データ')

#Define an empty variable.
df = pd.DataFrame(columns = [])
Start_Time_TNS_0=[]
Stop_Time_TNS_0=[]
Start_Time_TNS_1=[]
Stop_Time_TNS_1=[]
Start_Time_TNS_2=[]
Stop_Time_TNS_2=[]
Start_Time_TNS_3=[]
Stop_Time_TNS_3=[]

```

```python


#Read and extract CSV files.
for i in sorted(glob.glob("*.csv")):
    df= pd.read_csv(i)
    print("The loaded file will be" + i + ".")

    #Data is extracted according to the value of TNS.
    df_TNS_0=df[df["TNS"]==0]
    df_TNS_1=df[df["TNS"]==1]
    df_TNS_2=df[df["TNS"]==2]
    df_TNS_3=df[df["TNS"]==3]
    
    #Make sure the data is not empty.
    bool_TNS_0=not df_TNS_0.empty
    bool_TNS_1=not df_TNS_1.empty
    bool_TNS_2=not df_TNS_2.empty
    bool_TNS_3=not df_TNS_3.empty
    
    
    if bool_TNS_0:
        print("Since there is data with a TNS of 0, the process of extracting the start and end times of the scene is executed.")
        Time_TNS_0=df_TNS_0["TimeStamp"]#Extracting the value of the TimeStamp column.
        Start_Time_TNS_0.append(Time_TNS_0.iloc[0])#Extracting the value of the first row, TimeStamp column.。
        Stop_Time_TNS_0.append(Time_TNS_0.iloc[-1])#Extracting the value of the last row, TimeStamp column.。

    if bool_TNS_1:
        print("Since there is data with a TNS of 1, the process of extracting the start and end times of the scene is executed.")
        Time_TNS_1=df_TNS_1["TimeStamp"]#Extracting the value of the TimeStamp column.
        Start_Time_TNS_1.append(Time_TNS_1.iloc[0])#Extracting the value of the first row, TimeStamp column.
        Stop_Time_TNS_1.append(Time_TNS_1.iloc[-1])#Extracting the value of the last row, TimeStamp column.

        
    if bool_TNS_2:
        print("Since there is data with a TNS of 2, the process of extracting the start and end times of the scene is executed.")
        Time_TNS_2=df_TNS_2["TimeStamp"]#Extracting the value of the TimeStamp column.
        Start_Time_TNS_2.append(Time_TNS_2.iloc[0])#Extracting the value of the first row, TimeStamp column
        Stop_Time_TNS_2.append(Time_TNS_2.iloc[-1])#Extracting the value of the last row, TimeStamp column.
        
    if bool_TNS_3:
        print("Since there is data with a TNS of 3, the process of extracting the start and end times of the scene is executed.")
        Time_TNS_3=df_TNS_3["TimeStamp"]#Extracting the value of the TimeStamp column.
        Start_Time_TNS_3.append(Time_TNS_3.iloc[0])#Extracting the value of the first row, TimeStamp column
        Stop_Time_TNS_3.append(Time_TNS_3.iloc[-1])#Extracting the value of the last row, TimeStamp column.




```

```python
#Cut UnixTime from the video file name.


#Define an empty variable.
lists_Video_Filename=[]
lists_Video_UnixTime=[]
lists_Video_duration=[]
lists_Video_finish=[]

for i in sorted(glob.glob("*.mp4")):
    print("The file name before cutting is",end="")
    print(i)
    lists_Video_Filename.append(i)
    clip = VideoFileClip(i)
    lists_Video_duration.append(clip.duration)    
    i=i.replace("003_05_","")
    i=i.replace(".bin.mp4","")
    print("After cutting, UNIX Time is",end="")
    i=int(i)
    i=float(i)
    i=i/1000
    print(i)
    lists_Video_UnixTime.append(i)
    finish=i+clip.duration
    lists_Video_finish.append(finish)
    
    
print()
print("Below is a list of video file names.")
print(lists_Video_Filename)
print("The Unix Time included in the video file name is as follows (the date and time when the video started).")
print(lists_Video_UnixTime)
print("The length (seconds) of the video is as follows.")
print(lists_Video_duration)
print("The date and time when the video ends is as follows.")
print(lists_Video_finish)
```
